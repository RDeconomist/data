```python
import pandas as pd
import itertools
```

```python
df = pd.read_csv('/Users/finn/Documents/GitHub/data/e4e9/wealth_inequality.csv', delimiter=';', skiprows=1)
df.dropna(inplace=True)
```

```python
df.columns = ['Country', 'measure_long', 'Series', 'Year', 'Value']
df['ISO2'] = df['measure_long'].str[17:19]
df = df [['Country', 'ISO2', 'Series', 'Year', 'Value']]
```

```python
years_to_fill = list(range(1807, 2022))
countries = df.Country.unique()
country_codes = df.country_code.unique()
combined = [countries, ['padding'],['padding'],years_to_fill, [-1] , country_codes]
empty_df = pd.DataFrame(columns = df.columns, data=list(itertools.product(*combined)))
```

```python
iso_df = pd.read_csv('https://raw.githubusercontent.com/lukes/ISO-3166-Countries-with-Regional-Codes/master/all/all.csv')
df['country_code'] = df['ISO2'].map(iso_df.set_index('alpha-2')['country-code'])
# Manually fill Namibia to 516
df.loc[df['ISO2'] == 'NA', 'country_code'] = 516
df['country_code'] = df['country_code'].astype(int)
```

```python
empty_df = pd.DataFrame(columns=df.columns)
years_to_fill = list(range(1807, 2022))
countries = df.Country.unique()
country_codes = df.country_code.unique()
for i, country, country_code in zip(range(len(countries)), countries, country_codes):
    empty_df = empty_df.append(
    pd.DataFrame(
        [country]*len(years_to_fill), 
        ['padding']*len(years_to_fill), 
        ['padding']*len(years_to_fill), 
        years_to_fill, 
        country_code*len(years_to_fill)
    ),
    columns=df.columns
)
```
