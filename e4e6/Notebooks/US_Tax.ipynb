### Recessions

```python
NBER_df = pd.read_json("http://data.nber.org/data/cycles/business_cycle_dates.json")
NBER_df = NBER_df.iloc[1:,:]
```

```python
NBER_df
NBER_df.columns = ['Start', 'End']
```

```python
pre_NBER_range_recessions = [['17850615', '17881215'], ['17890615', '17931215'], ['17920615', '17921215'], ['17960615', '17991215'], ['18020615', '18041215'], ['18070615', '18101215'], ['18120615', '18121215'], ['18150615', '18211215'], ['18220615', '18231215'], ['18250615', '18261215'], ['18280615', '18291215'], ['18330615', '18341215'], ['18360615', '18381215'], ['18390615', '18431215'], ['18450615', '18461215'], ['18471001', '18481130'], ['18530601', '18541231']]
pre_df = pd.DataFrame(pre_NBER_range_recessions)
pre_df.columns = ['Start', 'End']
pre_df.Start = pre_df.Start.str[0:4] + '-' + pre_df.Start.str[4:6] + '-' + pre_df.Start.str[6:8]
pre_df.End = pre_df.End.str[0:4] + '-' + pre_df.End.str[4:6] + '-' + pre_df.End.str[6:8]
```

```python
recession_df = pd.concat([pre_df, NBER_df], axis=0)
```

```python
recession_df.Start = pd.to_datetime(recession_df.Start)
recession_df.End = pd.to_datetime(recession_df.End)
```
