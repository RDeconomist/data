```python
import pandas as pd
```

```python
df = pd.read_excel("chart4_7_baumol.xls")
dates = df.Date
df = df.iloc[1:,:]
base_year = df.iloc[0, 1:]
df = df.iloc[:, 1:].apply(lambda x: x / base_year * 100, axis=1)
df["Date"] = dates
df
df = df.melt(id_vars=["Date"], var_name="Series", value_name="Value")
df.to_csv("/Users/finn/Documents/GitHub/RADataHub/e4e/ch4_Money/data/chart4_7_baumol.csv", index=False)
```

```python

bases = {}
def reindex(x):
    global bases
    # reindex to 1980-01-01 = 100
    


    if x.Series not in bases:
        base = df.query("Date == '1980-01-01'")["Value"].values[0]
        bases[x.Series] = base
    else:  
        base = bases[x.Series]
    return x.Value / base * 100
df["Value"] = df.apply(reindex, axis=1)
```
