```python
import pandas as pd
import json
```

```python
df = pd.read_excel("NY_Chicago_Prices.xlsx")
df = df.melt(id_vars="Date", var_name="Series", value_name="Value")
df = df.dropna()
df.to_csv("/Users/finn/Documents/GitHub/RADataHub/e4e/ch4_Money/data/chart4_8_ny_ch_prices.csv", index=False)
```

```python
df
```

```python
# add a year column
df["Year"] = df["Date"].dt.year
df["Month"] = df["Date"].dt.month
with open("index_dict", "w") as f:
    d = df.query("(Series == 'NY_Index' | Series == 'Chicago_Index' ) & Month == 1")[['Year', "Value", "Series"]].to_dict(orient='records')
    d = {f"{d['Series']}_{d['Year']}": d["Value"] for d in d}
    json.dump(d, f)
```
